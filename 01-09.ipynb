```python
#practical 8
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000)
x = np.linspace(0, 2 * np.pi, 100)  # Data for sine/cosine plot
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.plot(x, y_sin, label='sin(x)', color='green')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#pract 9
import matplotlib.pyplot as plt
import numpy as np


data_matrix = np.random.randn(100, 5)  
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot histogram
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(data_matrix[:, 0], bins=20, color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.plot(x, y_sin, label='sin(x)', color='green')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()

plt.subplot(2, 2, 4)
for i in range(5):
    plt.plot(x, data_matrix[:, i], label=f'Data {i}', color=plt.cm.viridis(i / 4), linestyle='--')
plt.title('Multiple Line Plots')
plt.xlabel('x')
plt.ylabel('Data')
plt.legend()

plt.tight_layout()
plt.show()

```
